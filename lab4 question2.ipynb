```python
#question 2

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = 'https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv'
df = pd.read_csv(url)
feature_name = 'medv'
min_val = df[feature_name].min()
max_val = df[feature_name].max()
df['minmax'] = (df[feature_name] - min_val) / (max_val - min_val)
print(f"MinMax value for '{feature_name}':")
print(df['minmax'].describe())

plt.figure(figsize=(8, 6))
sns.histplot(df['minmax'], bins=20, kde=True, color='skyblue')
plt.xlabel('MinMax Value')
plt.ylabel('Frequency')
plt.title(f'Distribution of {feature_name} (MinMax)')
plt.show()
```
